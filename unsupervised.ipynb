```
# 1. Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 2. Define the dataset
data = {
    "hours_studied": [8, 2, 5, 1, 7, 3, 6, 4],
    "attendance": [90, 40, 75, 30, 85, 55, 80, 60]
}

df = pd.DataFrame(data)

# 3. Prepare features (NO labels needed for unsupervised learning)
X = df[["hours_studied", "attendance"]]

# 4. Apply KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# 5. Visualize the clusters
plt.figure(figsize=(8,6))
plt.scatter(df['hours_studied'], df['attendance'], c=df['cluster'], cmap='viridis', s=100)
plt.xlabel("Hours Studied")
plt.ylabel("Attendance (%)")
plt.title("Student Clusters Based on Study Hours and Attendance")
plt.grid(True)
plt.show()

# 6. See the clustering result
print(df)
```
